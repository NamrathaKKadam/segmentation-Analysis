```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Load your dataset
df = pd.read_csv(r"C:\Users\Namratha Kadam\OneDrive\Desktop\ifood_df.csv")  

# Selecting relevant features for segmentation
features = ['Income', 'Age', 'MntTotal', 'NumWebPurchases', 
            'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

df_segmentation = df[features].dropna()  # Removing missing values
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_segmentation)  # Transform to normalized values
```

```python
wcss = []  # Within-cluster sum of squares
silhouette_scores = []

for k in range(2, 11):  # Checking for clusters from 2 to 10
    kmeans = KMeans(n_clusters=k, random_state=42,n_init=10)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)  # Append WCSS values
    silhouette_scores.append(silhouette_score(df_scaled, kmeans.labels_))  # Silhouette Score

# Plot Elbow Method
plt.figure(figsize=(10,5))
plt.plot(range(2, 11), wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method - Optimal K")
plt.show()

# Plot Silhouette Scores
plt.figure(figsize=(10,5))
plt.plot(range(2, 11), silhouette_scores, marker='s', linestyle='--', color='red')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Different K Values")
plt.show()
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_segmentation['Cluster'] = kmeans.fit_predict(df_segmentation)

# Plot scatterplot
plt.figure(figsize=(10,6))
sns.scatterplot(x=df_segmentation['Income'], 
                y=df_segmentation['MntTotal'], 
                hue=df_segmentation['Cluster'], 
                palette="viridis")
plt.title("Customer Segmentation Based on Income & Total Spend")
plt.xlabel("Income")
plt.ylabel("Total Spend")
plt.show()

```

```python
cluster_summary = df_segmentation.groupby('Cluster').mean()
print(cluster_summary)
```
